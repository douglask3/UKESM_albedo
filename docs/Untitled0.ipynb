# Optimize allometry

## Allometry scheme in JULES

Height is calculated as a function of wood carbon $W$

\begin{equation}
h = \frac{W}{a_{ws} \cdot \eta_{sl}} \cdot \Big(\frac{a_{wl}}{W}\Big)^{1/b_{wl}}
\end{equation}

and

\begin{equation}
W = a_{wl} \\cdot LAI_{b}^{b_{wl}}
\end{equation}


where $\eta_{sl}$ is a constant of proportionality, set to 0.01 for all PFTs; where $LAI_{b}$ is the Lead Area Index when plant is in "full leaf" (i.e when $T_{c} > T_{off}$) and $a_{wl}$, $b_{wl}$ and $a_{ws}$ are PFT-dependant allometric parameters (see xxxx and Clark et al. 2011 for more details).

Here, $a_{wl}$, $b_{wl}$ and $a_{ws}$ are optimized against GC3.1 height and LAI ancil files (??source??)

### Optimize

Now to define the allometry equations

```python
def sim_woodC(LAI_b, a_wl, b_wl):
    return a_wl * LAI_b ** b_wl
    
def sim_height(LAI_b, a_wl, b_wl, a_ws = 1.0, eta_sl = 0.01):
    W = woodC(LAI_b, a_wl, b_wl)
    return W / (a_ws * eta_sl) * ((a_wl / W)** (1/b_wl))
```

This is the function that will be minimised for each PFT

```python

def minFun(params, pft):
    a_wl, b_wl, a_ws  = params
    h = sim_height(lai_b[pft].data, a_wl, b_wl, a_ws)  
    h -= height[pft].data
    h = np.absolute(h)
    return h.mean()
```

```python
def optimize_pft(pft):
    print(tile_name[pft])
    print(start[pft])
    res = minimize(minFun, start[pft], args = (pft), bounds = bounds, method='SLSQP').x
    print(res)
    return res

res = [optimize_pft(pft) for pft in range(0,9)]
```

```python
def construct_height(pft, npfts, levels = [0.1, 0.5, 1, 5, 10, 20, 50]):
    plt.figsize(15,2.5 * npft)
    a_wl1, b_wl1, a_ws1 = start[pft]
    a_wl2, b_wl2, a_ws2 = res[pft]    
    
    h0 = height[pft]
    h1 = h0.copy()
    h2 = h0.copy()
    
    h1.data = sim_height(lai_b[pft].data, a_wl1, b_wl1, a_ws1)
    h2.data = sim_height(lai_b[pft].data, a_wl2, b_wl2, a_ws2)
    
    hgt_c = plot_cube(h0, npft, 3, 3*pft + 1, 'brewer_YlOrBr_09', levels, 'max')
    if pfti == 0: plt.gca().text(0.5, 1.05, 'GC3.1', transform = plt.gca().transAxes)
    hgt_c = plot_cube(h1, npft, 3, 3*pft + 2, 'brewer_YlOrBr_09', levels, 'max')
    if pfti == 0: plt.gca().text(0.5, 1.05, 'UKESM0.9 params, transform = plt.gca().transAxes)
    hgt_c = plot_cube(h2, npft, 3, 3*pft + 3, 'brewer_YlOrBr_09', levels, 'max')
    if pfti == 0: plt.gca().text(0.5, 1.05, 'Optimized params', transform = plt.gca().transAxes)
    
    
    

def construct_height_pfts(pfts, *args, **kw):
    npfts = len(pfts)
    [construct_height(pft, npfts) for pft in pfts]
    
construct_height_pfts(range(0,5), levels = [1, 10, 15, 20, 21, 22, 23, 24])
```
